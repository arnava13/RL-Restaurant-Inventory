```python
import matplotlib.pyplot as plt
import numpy as np
import IPython

from src.env import KitchenEnv

env = KitchenEnv()

for i in range(10):
    env.step(env.action_space.sample())
    env.render()
    IPython.display.clear_output(wait=True)
    IPython.display.display(plt.gcf())
    plt.close()
```

```python
def render(KitchenEnv):
        max_raw = int(max(KitchenEnv.raw_shelf_life))
        max_prep = int(max(KitchenEnv.prep_shelf_life))

        raw_mat = np.zeros((KitchenEnv.n_raw, max_raw))
        prep_mat = np.zeros((KitchenEnv.n_prep, max_prep))

        for i in range(KitchenEnv.n_raw):
            raw_mat[i, : len(KitchenEnv.raw_inv[i])] = KitchenEnv.raw_inv[i]
        for j in range(KitchenEnv.n_prep):
            prep_mat[j, : len(KitchenEnv.prep_inv[j])] = KitchenEnv.prep_inv[j]

        plt.figure(figsize=(8, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(raw_mat, cmap="YlGn", aspect="auto", origin="lower")
        plt.title(f"Raw Inventory (Day {KitchenEnv.day})")
        plt.colorbar()

        plt.subplot(1, 2, 2)
        plt.imshow(prep_mat, cmap="YlOrBr", aspect="auto", origin="lower")
        plt.title("Prepared Inventory")
        plt.colorbar()

        plt.tight_layout()
        IPython.display.clear_output(wait=True)
        IPython.display.display(plt.gcf())
        plt.close()
```
